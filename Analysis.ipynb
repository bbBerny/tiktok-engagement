```python
import numpy as np
import matplotlib.pyplot as plt

# Define x range and parameters
x = np.linspace(0, 1, 400)
a_values = [1, 5, 10]
b = 0.5

# Create side-by-side subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot Logistic Sigmoid
for a in a_values:
    axes[0].plot(x, 1 / (1 + np.exp(-a * (x - b))), label=f"a={a}")
axes[0].set_title("Logistic Sigmoid (b=0)")
axes[0].set_xlabel("x")
axes[0].set_ylabel("sigmoid(x)")
axes[0].legend()

# Plot Scaled Tanh
for a in a_values:
    axes[1].plot(x, (np.tanh(a * (x - b)) + 1) / 2, label=f"a={a}")
axes[1].set_title("Scaled Tanh (b=0)")
axes[1].set_xlabel("x")
axes[1].set_ylabel("scaled_tanh(x)")
axes[1].legend()

plt.tight_layout()
plt.show()
```

Our choices for threshold, center and steepness must be grounded to our data structure and aligned with our research goals.

- We will look into the data and inspect its distribution, a good candidate for a center point `b` is an inflection point, also median, mean and a specific percentile like 75th makes sense
- We will play with these values and plot its soft scores in order to visually confirm our center `b`

After, our objective is to define a value `c` for the threshold

- We want to decide `c` accordingly to our objectives, we want to avoid false positives as much as possible so we will choose a higher 'c' cutoff

Note: If we were to work with `n` features and an output `y`

---

#### Defining steepness 'a'

We will apply steepness with two values:
- 0.1
- 5


```python
a_values = [0.1, 5]
```

---

#### Defining center 'b'

For the center we will be looking at the following data:
- Inflection point in the histogram -> This is a hyperparameter and it can be succeptible to experimentation
- Median
- Mean
- 75th percentile
- 25th percentile

```python
mean_b = df['Engagement_MinMaxScaled'].mean()
median_b = df['Engagement_MinMaxScaled'].median()
p75_b = df['Engagement_MinMaxScaled'].quantile(0.75)
p25_b = df['Engagement_MinMaxScaled'].quantile(0.25)
inflection_b = 0.05

print(f'The mean is: {mean_b}')
print(f'The median is: {median_b}')
print(f'The 75th percentile is: {p75_b}')
print(f'The 25th percentile is: {p25_b}')
print(f'The inflection point in the data is: {inflection_b}')
```

Since median and 25th percentile is zero, we will not use these values

```python
b_values = [mean_b, p75_b, inflection_b]
```

---

#### Defining threshold 'c'

For threshold values we will be trying to avoid type 1 errors (false-positives) so we will be careful not to label with 1 a case that is 0

For that reason our threshold values will be conservative

```python
c_values = [0.5, 0.6, 0.7]
```

---

With these parameters in mind we will go ahead and calculate the given columns for log-sig -> (2x3x3) -> 18 columns generated

```python
from scipy.special import expit

x = df['Engagement_MinMaxScaled']

for a in a_values:
    for b in b_values:
        scores = expit(a * (x - b))
        for c in c_values:
            df[f'Engagement_logsig_a{int(a*100):02d}_b{int(b*100):02d}_c{int(c*100):02d}'] = (scores > c).astype(int)
            #print(f'With values a: {a}, b: {b}, c: {c}')
            counts = df[f'Engagement_logsig_a{int(a*100):02d}_b{int(b*100):02d}_c{int(c*100):02d}'].value_counts(dropna = False)
            
            #print(f'The count of ones is: {counts.get(1)}')

df.head()
```

By looking at the dataframe we can see some repeated columns and also some columns with just one value, we will go ahead and delete those

```python
# 1) Drop any column that never changes (constant)
df = df.loc[:, df.nunique(dropna=False) > 1]

# 2) Drop any columns whose entire contents are identical to an earlier one
df = df.loc[:, ~df.T.duplicated(keep="first")]
```

#### Step 5

- We've now turned our continuous engagement signal into a handful of binary series using log-sig

- We want to ask: Is there really engagement here?

- Null hypothesis: The true probability of engagement is no greater than some baseline `p`
- Alternative hypothesis: The true probability of engagement is greater than `p`


##### Definition of p (null proportion) for TikTok engagement

To perform a test of proportions (binomial or z) we need to define a reference value , which represents the proportion of engaged segments under the null hypothesis (no effect). From the study by Garca-Marn & Salvat-Martinrey (2022) on fact-checkers in TikTok we can extract:

1. Average engagement rate.

- The engagement rate () has a mean of 7.51% and standard deviation of 6.25%

- In scale [0,1] this is equivalent to 0.0751.

2. Normalized Min-Max scale

- When the engagement rate is normalized to the range [0,1], the resulting average is 0.250875.

Thus we will propose p as 0.250875


```python
from scipy.stats import binomtest

p0 = 0.250875

for col in df.filter(like='Engagement_logsig'):
    n1 = df[col].sum()      # number of ones
    N = len(df[col])        # total trials

    # test H0: p = p0 vs H1: p > p0
    pval = binomtest(n1, N, p=p0, alternative='greater')
    print(col, 'n1/N =', f'{n1}/{N}', 'p-value =', pval)
```
