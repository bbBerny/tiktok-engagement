```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler

# 1. Load your data
df = original_df

# 2. Identify columns
cont_cols = list(df.columns[3:56])   # continuous features
disc_cols = list(df.columns[56:93])  # discrete (0/1) features

# 3. Choose neighbour features
neigh = []
if 'GSR Raw' in df.columns:
    neigh.append('GSR Raw')

# 4. Build the matrix to impute
impute_cols = cont_cols + disc_cols + neigh
X = df[impute_cols].copy()

# 6. Scale everything
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 7. KNN impute
imputer = KNNImputer(n_neighbors=5, weights='uniform')
X_filled = imputer.fit_transform(X_scaled)

# 8. Inverse‐scale
X_inv = scaler.inverse_transform(X_filled)
X_imputed = pd.DataFrame(X_inv, columns=impute_cols, index=df.index)

# 9. Put imputed values back into your original df
df_imputed = df.copy()

#   – continuous: use as is
df_imputed[cont_cols] = X_imputed[cont_cols]

#   – discrete: round to nearest integer (0 or 1) and clip
df_imputed[disc_cols] = (
    X_imputed[disc_cols]
      .round()
      .clip(lower=0, upper=1)
      .astype(int)
)

df_imputed

# Now df_imputed has your KNN‐filled features
```
